## load `train.csv` data with pandas

```python
import pandas as pd
import numpy as np

z = pd.read_csv("train.csv")
print ("shape of loaded dataframe", z.shape)
```

#Limpieza y desclasificación de variables

dropear ID

```python
z.drop('ID', axis=1, inplace=True)
```

```python
z.head()
```

convertir ESTU_PRGM_ACADEMICO y ESTU_PRGM_DEPARTAMENTO, FAMI_EDUCACIONPADRE, FAMI_EDUCACIONMADRE, FAMI_TIENEINTERNET, FAMI_ESTRATOVIVIENDA, ESTU_PAGOMATRICULAPROPIO valores numericos:

```python
# Assign unique numeric values to each unique program
def cleaning_variables(dataframe):
  dataframe["ESTU_PRGM_ACADEMICO"] = dataframe["ESTU_PRGM_ACADEMICO"].astype("category").cat.codes
  dataframe["ESTU_PRGM_DEPARTAMENTO"] = dataframe["ESTU_PRGM_DEPARTAMENTO"].astype("category").cat.codes
  dataframe["FAMI_EDUCACIONPADRE"] = dataframe["FAMI_EDUCACIONPADRE"].astype("category").cat.codes
  dataframe["FAMI_EDUCACIONMADRE"] = dataframe["FAMI_EDUCACIONMADRE"].astype("category").cat.codes
  dataframe["FAMI_TIENEINTERNET"] = dataframe["FAMI_TIENEINTERNET"].astype("category").cat.codes
  dataframe["FAMI_ESTRATOVIVIENDA"] = dataframe["FAMI_ESTRATOVIVIENDA"].astype("category").cat.codes
  dataframe["ESTU_PAGOMATRICULAPROPIO"] = dataframe["ESTU_PAGOMATRICULAPROPIO"].astype("category").cat.codes
# Display the DataFrame
cleaning_variables(z)
```

```python
z.head()
```

Limpieza para ESTU_VALORMATRICULAUNIVERSIDAD




```python
def limpieza_valormatricula(dataframe):
  # Asignar valores únicos a cada categoría (ignorando NaN)
  dataframe['ESTU_VALORMATRICULAUNIVERSIDAD'] = pd.factorize(dataframe['ESTU_VALORMATRICULAUNIVERSIDAD'])[0]

  # Calcular la media de los valores numéricos (ignorando NaN)
  media_valor = dataframe['ESTU_VALORMATRICULAUNIVERSIDAD'].replace(-1, np.nan).mean()

  # Reemplazar los valores NaN con la media
  dataframe['ESTU_VALORMATRICULAUNIVERSIDAD'] = dataframe['ESTU_VALORMATRICULAUNIVERSIDAD'].replace(-1, media_valor)
limpieza_valormatricula(z)
```

```python
z.head()
```

Limpieza para ESTU_HORASSEMANATRABAJA

```python
import pandas as pd
import numpy as np

def limpiar_horas(dataframe):
    # Definir un diccionario con los rangos y sus valores representativos
    valores_map = {
        'Menos de 10 horas': 5,         # Promedio de 0 a 10 horas
        'Entre 11 y 20 horas': 15,      # Promedio de 11 a 20 horas
        'Entre 21 y 30 horas': 25,      # Promedio de 21 a 30 horas
        'Más de 30 horas': 35,          # Asignar valor arbitrario de 35 horas
        '0': 0                          # No trabaja
    }

    # Mapear las categorías de texto a los valores numéricos
    dataframe['ESTU_HORASSEMANATRABAJA'] = dataframe['ESTU_HORASSEMANATRABAJA'].map(valores_map)

    # Convertir los valores a tipo numérico, manejando los errores como NaN
    dataframe['ESTU_HORASSEMANATRABAJA'] = pd.to_numeric(dataframe['ESTU_HORASSEMANATRABAJA'], errors='coerce')

    # Calcular la media y desviación estándar de los valores diferentes a 0 para evitar sesgo
    mean_value = dataframe.loc[dataframe['ESTU_HORASSEMANATRABAJA'] > 0, 'ESTU_HORASSEMANATRABAJA'].mean()
    std_dev = dataframe.loc[dataframe['ESTU_HORASSEMANATRABAJA'] > 0, 'ESTU_HORASSEMANATRABAJA'].std()

    # Contar los valores NaN para imputar los valores faltantes
    missing_values_count = dataframe['ESTU_HORASSEMANATRABAJA'].isna().sum()

    # Sustituir los NaN generando valores aleatorios basados en una distribución normal
    dataframe.loc[dataframe['ESTU_HORASSEMANATRABAJA'].isna(), 'ESTU_HORASSEMANATRABAJA'] = np.random.normal(mean_value, std_dev, missing_values_count)

    return dataframe
limpiar_horas(z)
z.head()
```

# convert target (prediction) column into discrete values

```python
y_col = 'RENDIMIENTO_GLOBAL'

rmap = {'alto': 3, 'bajo':0, 'medio-bajo':1, 'medio-alto':2}
z[y_col] = [rmap[i] for i in z[y_col]]
z.head()
```

#INICIO DEL MODELO SOLUCIÓN PARA ESTE ENTREGABLE (99)

build X and y for training a model
observe we sort columns to make sure we always get the same ordering

```python
z = z[sorted(z.columns)]

X = z[[c for c in z.columns if c!=y_col]].values
y = z[y_col].values
X.shape, y.shape
```

split into train and test

```python

from sklearn.model_selection import train_test_split
Xtr, Xts, ytr, yts = train_test_split(X,y, train_size=0.8)
Xtr.shape, Xts.shape, ytr.shape, yts.shape
```

Make a random forest for classification
train model


```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, min_samples_split=2, min_samples_leaf=4, max_features= "sqrt", max_depth=20 ,random_state=42 )
rf.fit(Xtr, ytr)
```

predicciones locales

```python
preds_tr = rf.predict(Xtr)
preds_ts = rf.predict(Xts)
```

```python
np.mean(preds_tr==ytr), np.mean(preds_ts==yts)
```
